```python
import numpy as np
from scipy.stats import rv_continuous

class CustomDistribution(rv_continuous):
    def _pdf(self, x): #  переопределение метода
        return 3*x**2*np.exp(-x**3)*(x >= 0)

custom_distr = CustomDistribution(name='custom_distr', a=0)

small_n = custom_distr.rvs(size=10)
big_n = custom_distr.rvs(size=1000)
print(f'случайные 1000 значений из распределения: {big_n}')
print(f'случайные 10 значений из распределения: {small_n}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

big_n = custom_distr.rvs(size=1000)

def ecdf(data):
  
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return x, y

x, y = ecdf(big_n)


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.hist(big_n, bins=30, density=True, alpha=0.5, color='blue')
plt.title('Гистограмма')
plt.xlabel('Значения')
plt.ylabel('Плотность')


plt.subplot(1, 2, 2)
plt.plot(x, y, marker='.', linestyle='none', color='red')
plt.title('Эмпирическая Кумулятивная Функция Распределения')
plt.xlabel('Значения')
plt.ylabel('ECDF')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Функция, возвращающая обратную функцию распределения
def inverse_cdf(y):
    return (-np.log(1 - y))**(1/3)


def generate_random_numbers(size):
    uniform_random_numbers = np.random.uniform(0, 1, size)
    return inverse_cdf(uniform_random_numbers)

random_numbers = generate_random_numbers(1000)
print(random_numbers)
```
